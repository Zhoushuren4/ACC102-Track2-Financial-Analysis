```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
ACC102 Track 2 Assignment
Financial Ratio Analysis for S&P 500 Companies
Author: [Your Name]
ID: [Your Student ID]
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Use a sample financial dataset (simplified for demonstration)
data = {
    'Company': ['Apple', 'Microsoft', 'Amazon', 'Google', 'Tesla', 
                'Johnson & Johnson', 'Walmart', 'JP Morgan', 'Visa', 'Nike'],
    'Current_Ratio': [1.12, 1.87, 1.05, 1.63, 1.23, 1.45, 0.82, 1.01, 1.56, 2.10],
    'Debt_to_Equity': [1.20, 0.45, 1.32, 0.38, 1.89, 0.56, 0.73, 1.54, 0.41, 0.67],
    'ROE': [28.1, 38.5, 15.3, 25.7, 18.2, 22.4, 17.8, 12.1, 32.5, 29.3],
    'Profit_Margin': [24.5, 36.7, 5.8, 21.3, 15.4, 18.2, 2.3, 22.1, 51.2, 12.9]
}

df = pd.DataFrame(data)

print("=== S&P 500 Financial Dataset ===")
print(df.head())
print("\n=== Dataset Information ===")
print(df.describe())

# Check missing values
print("\n=== Missing Values Check ===")
print(df.isnull().sum())

# No missing values in this dataset

print("\n=== Top 5 Companies by ROE ===")
print(df.sort_values(by='ROE', ascending=False).head())

print("\n=== Average Financial Ratios ===")
print(df[['Current_Ratio', 'Debt_to_Equity', 'ROE', 'Profit_Margin']].mean())

plt.style.use('seaborn-v0_8-white')

# Plot 1: ROE Comparison
plt.figure(figsize=(12,6))
plt.bar(df['Company'], df['ROE'], color='#2E86AB')
plt.title('Return on Equity (ROE) for S&P 500 Companies', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.ylabel('ROE (%)')
plt.tight_layout()
plt.savefig('roe_comparison.png', dpi=300)

# Plot 2: Profit Margin vs Current Ratio
plt.figure(figsize=(10,6))
plt.scatter(df['Current_Ratio'], df['Profit_Margin'], s=120, color='#A23B72', alpha=0.7)
for i, txt in enumerate(df['Company']):
    plt.annotate(txt, (df['Current_Ratio'][i], df['Profit_Margin'][i]))
plt.title('Profit Margin vs Current Ratio', fontsize=14)
plt.xlabel('Current Ratio')
plt.ylabel('Profit Margin (%)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('profit_vs_liquidity.png', dpi=300)

print("\nAnalysis completed successfully! Charts saved as PNG files.")
```
